```python
#guardo en una variable los pozos no convencionales y de la cuenca neuquina

proyecto_fopea_neuquina_noconvencional=proyecto_fopea.loc[(proyecto_fopea['cuenca'] == 'NEUQUINA') & (proyecto_fopea['tipo_de_recurso'] == 'NO CONVENCIONAL')]
proyecto_fopea_neuquina_noconvencional

#.loc selecciona las filas que cumplen la condición booleana
```

```python
#buscando cantidad de pozos unicos y no convencional por empresas descendente

pozos_empresas=proyecto_fopea_neuquina_noconvencional.groupby('idempresa')['idpozo'].nunique().sort_values(ascending=False)
pozos_empresas
```

```python
#calculo el total de pozos
total_pozos=pozos_empresas.sum()
print(f'Hay un total de {total_pozos} pozos unicos NO CONVENCIONALES de la cuenca Neuquina en 2025')
```

```python
#creo una columna para saber cuales son argentinas y cuales extrangeras

empresas_argentinas = [
    "YPF","PCR","PLU","ACO","OILS","BEN","PSD","CHE","QUIN","APS",
    "TAU","TPT","VST","PAL","PPSA","GTIE","MSA","PAM","EME","PAMP",
    "PES","ROC","PET","HAT","SJPS","Z001","VNO","ENE1","PSM","PTRE",
    "ISIM","PPAM","EMEA","PCN","AESA"
] #lista de empresas argentinas

proyecto_fopea_neuquina_noconvencional.loc[:, 'origen'] = np.where(proyecto_fopea_neuquina_noconvencional['idempresa'].isin(empresas_argentinas),'arg','ext')
#.loc cree una columna llamada origen en todas las filas
#funcion np.where(es como IF en numpy)
#compara los valores de la lista empresas_argentinas con los de la columna idempresa,si es verdadero arg sino ext

proyecto_fopea_neuquina_noconvencional
```

```python
#veo cuantos pozos convencionales de empresas argentinas y extranjeras hay en el df

proyecto_fopea_neuquina_noconvencional.groupby('origen')['idpozo'].nunique()
```

# TRABAJANDO CON ARREGLOS

```python
#Creando un arreglo
arr_origenes = proyecto_fopea_neuquina_noconvencional['origen'].to_numpy() #.to_numpy() convierte una columna en un array en este caso la columna origen
arr_prod_gas = proyecto_fopea_neuquina_noconvencional['prod_gas'].to_numpy()
arr_origenes
arr_prod_gas
```

```python
#Informacion de los arreglos
print('arreglo origenes')
print(arr_origenes)
print('Dimension: ',arr_origenes.ndim)
print('Tamaño: ', arr_origenes.size)
print('Tipo de dato: ', arr_origenes.dtype)
print('-----------------------------')
print('arreglo prod_gas')
print(arr_prod_gas)
print('Dimension: ',arr_prod_gas.ndim)
print('Tamaño: ', arr_prod_gas.size)
print('Tipo de dato: ', arr_prod_gas.dtype)
```

```python
#contando cuantos registros hay en origenes segun los size del arreglo

arr_origenes_arg=arr_origenes[arr_origenes=='arg']# == booleano de NumPy para seleccionar solo los valores 'arg' del arr_origenes lo guardo en arr_origenes_arg
print('Registros de origen argentinos: ',arr_origenes_arg.size)
arr_origenes_ext=arr_origenes[arr_origenes=='ext']
print('Registros de origen extranjeros: ',arr_origenes_ext.size)
```

```python
#Funcion para resumen estadistico de arreglos

def resumen_estadistico(arr): #definiendo una funcion que se llama resumen_estadistico y recibe un arreglo
    media = np.mean(arr) #busca la media de los datos
    mediana = np.median(arr) #busca la mediana
    desviacion = np.std(arr) #busca la desviacion
    percentiles = np.percentile(arr, [25, 50, 75]) #busca los percentiles
    return {
        'media': media,
        'mediana': mediana,
        'desviacion': desviacion,
        'percentiles': percentiles
           }
```

```python
#probando la funcion

resumen_prod = resumen_estadistico(arr_prod_gas) #guarda en resumen_prod el resultado de la def, le di el arreglo: arr_prod_gas
print(resumen_prod)
```

```python
valores = [suma_YPF, resto_argentinas, suma_extranjero]
colores = ['#FF9800', '#4CAF50', '#2196F3']
nombres = ['YPF', 'NAC', 'EXT']

fig, ax = plt.subplots(figsize=(8,2))

# Barra apilada
izquierda = 0
y_text = 0.6  # altura para los textos fuera de la barra

for valor, color, nombre in zip(valores, colores, nombres):
    # Dibuja la barra
    ax.barh(y=0, width=valor, left=izquierda, color=color, height=0.5)

    # Texto fuera de la barra, centrado sobre el segmento
    ax.text(izquierda + valor/2, y_text, f'{nombre}: {valor:,.0f}',
            ha='center', va='bottom', fontsize=9, fontweight='bold', color=color)

    izquierda += valor

# Ajustes visuales
ax.set_xlim(0, sum(valores))
ax.set_ylim(-0.5, 1)  # espacio arriba para los textos
ax.set_yticks([])
ax.set_xticks([])
for spine in ['top','right','left','bottom']:
    ax.spines[spine].set_visible(False)

ax.set_title('Producción Total de Gas ', pad=10)
plt.tight_layout()
plt.show()
```
